##Atlas de Recursos Digitales para la Literatura y el Teatro: Un Análisis de Bibliotecas Virtuales y Colecciones de Textos

#Descripción de la base de datos:
El conjunto de datos proporciona información sobre diversas bibliotecas virtuales y colecciones de textos, enfocándose en recursos digitales relacionados con la literatura y el teatro. Los datos incluyen el nombre de cada recurso, su URL, tipo de dato, formato, tipo de proyecto, responsable, año de comienzo, rango de edición, época, género textual, subgénero, autor, disponibilidad para descarga de texto completo y otros datos relevantes.

### Variables Críticas Identificadas:

1. **Nombre** (Variable categórica): Esta variable representa el nombre de la biblioteca virtual o colección de textos. Es importante para identificar y diferenciar entre los diferentes recursos.

2. **Tipo de proyecto** (Variable categórica): Esta variable indica el tipo de proyecto al que pertenece cada recurso, como "biblioteca digital" o "colección de textos". Es crucial para entender el enfoque y propósito de cada recurso.

3. **Año de comienzo** (Variable numérica): Representa el año en que se inició el proyecto. Es una variable clave para analizar la antigüedad y evolución de los recursos digitales.

4. **Época** (Variable categórica): Esta variable indica el periodo histórico al que pertenece el contenido de los textos. Es esencial para entender el contexto temporal de los recursos.

5. **Género textual** (Variable categórica): Define el género literario al que pertenece el texto, como "literatura" o "teatro". Es fundamental para analizar el tipo de contenido que ofrecen los recursos.

6. **Descarga full-text** (Variable categórica): Indica si el texto completo está disponible para descarga ("sí" o "no"). Es una variable crítica para evaluar el acceso que los usuarios tienen al contenido completo de los recursos.

### Hallazgos Interesantes:

1. **Evolución de Proyectos Digitales**: La base de datos revela una notable expansión en la creación de proyectos digitales relacionados con la literatura y el teatro desde la década de 1970, con un incremento significativo a partir del año 2000. Esto sugiere una creciente importancia de la digitalización en la preservación y difusión de textos literarios.

2. **Acceso a Texto Completo**: Sorprendentemente, más del 40% de los recursos no permiten la descarga completa del texto, lo que podría limitar el acceso del público a estos valiosos contenidos. Esto abre la puerta a discusiones sobre derechos de autor, acceso abierto y la democratización del conocimiento.

3. **Diversidad de Épocas y Géneros**: Los recursos abordan una amplia gama de periodos históricos y géneros literarios, destacándose la versatilidad de estos proyectos. Sin embargo, una gran parte de ellos cubre "varios" periodos, lo que sugiere un enfoque enciclopédico o multi-temático en lugar de uno especializado.

4. **Dominio de Colecciones de Textos**: Las colecciones de textos superan en número a las bibliotecas digitales, lo que indica una tendencia hacia la agrupación y edición crítica de obras específicas, posiblemente con un enfoque académico.



```python
file_path = 'https://raw.githubusercontent.com/morethanbooks/Atlas-de-Datos/master/atlas%20de%20datos.csv'
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(file_path)
data.head()
```

#1 ¿Cómo ha evolucionado el acceso a recursos literarios digitales en términos de disponibilidad de texto completo desde el inicio de estos proyectos hasta la actualidad?



```python

data['Década de Comienzo'] = (data['Año de comienzo'] // 10) * 10

descarga_por_decada = data.groupby('Década de Comienzo')['Descarga full-text'].apply(lambda x: (x == 'sí').mean() * 100)

descarga_por_decada.plot(kind='bar', color='skyblue')
plt.title('Porcentaje de Proyectos con Descarga Completa por Década')
plt.xlabel('Década de Comienzo')
plt.ylabel('Porcentaje de Proyectos con Descarga Completa')
plt.xticks(rotation=45)
plt.show()
```

#Conclusión:
El análisis muestra que la disponibilidad de descarga completa de textos ha disminuido ligeramente en los proyectos digitales más recientes en comparación con los más antiguos.

#Regresión Numérica (#9)
#¿Cómo influye el año de comienzo en la cantidad de datos normalizada de los proyectos digitales, y cómo podemos predecir la cantidad de datos normalizada para un proyecto con un año de comienzo específico?
#Formula: Cantidad de datos normalizada=β0 + β1 * Año de comienzo



```python
import pandas as pd
import statsmodels.api as sm

data_valid = data[['Año de comienzo', 'Cantidad de datos normalizada']].dropna()

X = data_valid['Año de comienzo']
y = data_valid['Cantidad de datos normalizada']

X = sm.add_constant(X)

modelo = sm.OLS(y, X).fit()

print(modelo.summary())

intercepto = modelo.params['const']
coef_anio_comienzo = modelo.params['Año de comienzo']
print(f"Fórmula de la regresión: Cantidad de datos normalizada = {intercepto:.4f} + {coef_anio_comienzo:.4f} * Año de comienzo")


anio_comienzo_nuevo = 2010
cantidad_datos_normalizada_predicha = intercepto + coef_anio_comienzo * anio_comienzo_nuevo

print(f"Predicción: Para un proyecto con año de comienzo en {anio_comienzo_nuevo}, la cantidad de datos normalizada predicha es {cantidad_datos_normalizada_predicha:.2f}")
```

#Conclusión:
Este modelo de regresión lineal permite predecir la cantidad de datos normalizada en función del año de comienzo de un proyecto digital. Se puede usar esta fórmula para hacer predicciones para diferentes años de comienzo, lo que permitirá analizar cómo el tiempo afecta la acumulación de datos en los proyectos.
