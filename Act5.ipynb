```python
import pandas as pd

file_path = 'https://raw.githubusercontent.com/Pabloblanco00/1/main/antropometria-dataset-1.csv'
data = pd.read_csv(file_path)

data.head()
```

```python
data.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(8, 5))
sns.boxplot(x='sexo', y='edad', data=data)
plt.title('Boxplot de edad')
plt.xlabel('Sexo')
plt.ylabel('edad')
plt.show()
```

```python
ndata = pd.pivot_table(data,
                              index = 'desc_ent',
                              columns= 'sexo',
                              aggfunc = 'median',
                              margins = True,
                              values = 'edad')
ndata = ndata.drop('All')
ndata
```

```python
mdata = ndata[2]
plt.figure(figsize=(8, 5))
plt.boxplot(mdata.dropna(), vert=False)
plt.title('Boxplot edad')
plt.ylabel('edad')
plt.grid(True)
plt.show()
```

```python
ndata = pd.pivot_table(data,
                              index = 'desc_ent',
                              columns= 'sexo',
                              aggfunc = 'median',
                              margins = True,
                              values = 'peso')
ndata = ndata.drop('All')
ndata
```

```python
ndata = data[data['sexo'] == 1].groupby('desc_ent')['cadera'].median()

plt.figure(figsize=(8, 5))
ndata.plot(kind='line', marker='o', color='red')
plt.title('Line Plot Cadera for Sex 1')
plt.xlabel('Region')
plt.ylabel('Median Cadera (cm)')
plt.xticks(rotation=90)
plt.grid(True)

plt.show()
```

```python
ndata = data[data['sexo'] == 2].groupby('desc_ent')['ropa'].sum()
tdata = ndata.sort_values(ascending=False).head(10)

plt.figure(figsize=(8, 5))
tdata.plot(kind='barh', color='green')
plt.title(' ropa for Sex 1')
plt.xlabel('Sum of Ropa')
plt.ylabel('Region')
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(data['talla'], bins=25, color='gray', edgecolor='black')
plt.title('Histogram of Talla')
plt.xlabel('Talla (cm)')
plt.ylabel('Frequency')
plt.grid(True)

plt.show()
```
